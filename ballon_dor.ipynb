```python
#loading important and used libraries 
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#loading the file 
ballon_dor = pd.read_csv(r"C:\Users\Aditya Rawat\Desktop\Ballon Dor.csv")
```

```python
ballon_dor.head()
```

```python
#AGE ANALYSIS
# average age
avg_age = ballon_dor["Age"].mean()
print(f"The average ballon dor winning age is: {avg_age}")
```

The average age of Ballon dor winner is around 29 years, suggesting that players typically reach peak individual performance in their late twenties.

```python
#youngest winner
youngest = ballon_dor.loc[ballon_dor["Age"].idxmin()]   #.idxmin() finds the row and .loc[] shows full player data
print(f"The youngest player to win ballon dor is {youngest["Winner"]}, he won it at {youngest["Age"]} years of age.")
```

```python
row_oldest = ballon_dor["Age"].idxmax()
oldest_dict = ballon_dor.loc[row_oldest]
print(f"The oldest player to win ballon dor is {oldest_dict["Winner"]}, he won it at {oldest_dict["Age"]} years of age.")
```

although the average ballon dor winning age is 29 but this shows that it can be won even in early twenties and late thirties 

```python
# histogram of ages
data = ballon_dor["Age"]
plt.hist(data, bins = range(22,38), edgecolor = "black", color = "g")
plt.title("Distribution of ballon dor winning ages")
plt.xlabel("Age")
plt.ylabel("The number of winners")
plt.show()
```

```python
#POSITION BIAS ANALYSIS
#who dominates?
position_count = ballon_dor["Position "].value_counts()
position_count
```

```python
# bar plot for position vs wins 
position_count = ballon_dor["Position "].value_counts()
plt.bar(position_count.index, position_count.values, edgecolor = "black", color = "r")
plt.title("Position vs Number of wins")
plt.xlabel("Positions")
plt.ylabel("Number of wins")
plt.show()
```

```python
#TOTAL TROPHIES INFLUENCE 
# creating a new column- Total_trophy_count
ballon_dor["Total_trophy_count"] = ballon_dor["Club_trophy_count"] + ballon_dor["International_trophy_count"]
```

```python
# displaying table
ballon_dor.head()
```

```python
#count of total_trophy_count:
total_trophy = ballon_dor["Total_trophy_count"].value_counts().sort_index()
total_trophy
```

```python
plt.scatter(total_trophy.index, total_trophy.values, c = "r", s = 50)
plt.title("Total trophies vs Number of ballon dor wins")
plt.xlabel("Total number of trophies")
plt.ylabel("Number of ballon dor winners")
plt.show()
```

```python
#INTERNATIONAL TROPHY EFFECT
#count of winners with and without international trophies 
international_trophy = ballon_dor["International_trophy_count"].value_counts()
international_trophy 
```

```python
#bar plot of winners with/without international trophies: 
no_of_int_trophy = international_trophy.index.map({0: "ZERO", 1: "ONE"})
no_of_winners = international_trophy.values
plt.bar(no_of_int_trophy, no_of_winners, edgecolor = "black", color = "blue")
plt.title("Number of international trophies vs Number of winners")
plt.xlabel("Number of international trophies")
plt.ylabel("Number of winners")
plt.show()
```

```python
#LEAGUE DOMINANCE
#counting number of winners from each league:
league_series = ballon_dor["League "].value_counts()
league_series
```

```python
#plotting pie chart
league = league_series.index
no_of_winners = league_series.values
plt.pie(no_of_winners, labels = league, autopct = "%0.01f%%")
plt.title("Winners by league distribution")
plt.show()
```

```python
#CLUB DOMINANCE 
#count of winners that clubs have produced:
club_series = ballon_dor["Club"].value_counts()
club_series
```

```python
#Bar plot showing clubs vs number of winners they have produced:
club = club_series.index
winners = club_series.values
plt.figure(figsize=(12,6))
plt.bar(club, winners, edgecolor = "black", color = "yellow")
plt.title("Number of winners produced by each club")
plt.xlabel("Clubs")
plt.ylabel("Number of winners")
plt.grid(True)
plt.show()
```

```python
#ERA DETECTION
ballon_dor[["Year", "Winner"]]
```

```python
#There are three eras:
# Pre duopoly era (2005-2007)
# Messi-Ronaldo duopoly era (2008-2017)
# Post duopoly era (2018-present)
ballon_dor["Era"] = ""
ballon_dor.loc[(ballon_dor["Year"]<=2007) & (ballon_dor["Year"]>=2005), "Era"] = "Pre-duopoly"
ballon_dor.loc[(ballon_dor["Year"]<=2017) & (ballon_dor["Year"]>=2008), "Era"] = "Duopoly"
ballon_dor.loc[ballon_dor["Year"]>=2018, "Era"] = "Post-duopoly"
```

```python
ballon_dor["Era"].value_counts()
```

```python
#pie chart of the eras:
era_series = ballon_dor["Era"].value_counts()
era = era_series.index
year = era_series.values
plt.pie(year, labels = era, autopct = "%0.01f%%")
plt.title("Era distribution")
plt.show()
```

```python
winner_codes, uniques = pd.factorize(ballon_dor["Winner"])
plt.figure(figsize = (12,5))
plt.yticks(range(len(uniques)), uniques)
plt.plot(ballon_dor["Year"], winner_codes, marker = 'o')
plt.xlabel("Year")
plt.ylabel("Winner Code")
plt.title("Ballon dor winners timeline")
plt.show()
```
